## Definición de modelo

```python
class Encoder(nn.Module):
  def __init__(self):
    super(Encoder, self).__init__()
    self.funcion = nn.ReLU()
    self.max = nn.MaxPool2d(kernel_size=2, stride=2)
    #256x256

    self.conv11 = nn.Conv2d(in_channels=3, out_channels=64, kernel_size=4, stride = 1,padding=2)
    self.conv12 = nn.Conv2d(in_channels=64, out_channels=64, kernel_size=4, stride = 1,padding=1)
    #128x128
    self.conv21 = nn.Conv2d(in_channels=64, out_channels=64*2, kernel_size=4, stride = 1,padding=2)
    self.conv22 = nn.Conv2d(in_channels=64*2, out_channels=64*2, kernel_size=4, stride = 1,padding=1)
    #64x64
    self.conv31 = nn.Conv2d(in_channels=64*2, out_channels=64*4, kernel_size=4, stride = 1,padding=2)
    self.conv32 = nn.Conv2d(in_channels=64*4, out_channels=64*4, kernel_size=4, stride = 1,padding=1)
    #32x32    
    self.conv41 = nn.Conv2d(in_channels=64*4, out_channels=64*8, kernel_size=4, stride = 1,padding=2)
    self.conv42 = nn.Conv2d(in_channels=64*8, out_channels=64*8, kernel_size=4, stride = 1,padding=1)
    #16x16
    self.conv51 = nn.Conv2d(in_channels=64*8, out_channels=64*16, kernel_size=3, stride = 1,padding=1)
   

  def forward(self, image):
    capas = []

    c1 = self.conv11(image)
    c1 = self.funcion(c1)
    c1 = self.conv12(c1)
    c1 = self.funcion(c1)

    p1 = self.max(c1)

    c2 = self.conv21(p1)
    c2 = self.funcion(c2)
    c2 = self.conv22(c2)
    c2 = self.funcion(c2)

    p2 = self.max(c2)

    c3 = self.conv31(p2)
    c3 = self.funcion(c3)
    c3 = self.conv32(c3)
    c3 = self.funcion(c3)
    p3 = self.max(c3)

    c4 = self.conv41(p3)
    c4 = self.funcion(c4)
    c4 = self.conv42(c4)
    c4 = self.funcion(c4)

    p4 = self.max(c4)

    capas.append(c4)
    capas.append(c3)
    capas.append(c2)
    capas.append(c1)
    c5 = self.conv51(p4)
    c5 = self.funcion(c5)
    #print("c5",c5.size())
    z=c5
    return z, capas

class Decoder(nn.Module):
  def __init__(self):
    super(Decoder, self).__init__()
    #self.fc    = nn.Linear(in_features=10,out_features=2*64*7*7)
    #self.fc    = nn.Linear(in_features=40,out_features=2*64*64*64)
    #self.convTran1 = nn.ConvTranspose2d(in_channels=2*64,out_channels=64, kernel_size=4, stride=2, padding=1)
    self.conv11 = nn.Conv2d(in_channels=16*64, out_channels=64*16, kernel_size=3, stride = 1,padding=1)
    self.convTran1 = nn.ConvTranspose2d(in_channels=16*64,out_channels=64*8, kernel_size=2, stride=2, padding=0)
    self.conv21 = nn.Conv2d(in_channels=16*64, out_channels=64*8, kernel_size=4, stride = 1,padding=2)
    self.conv22 = nn.Conv2d(in_channels=8*64, out_channels=64*8, kernel_size=4, stride = 1,padding=1)
    self.convTran2 = nn.ConvTranspose2d(in_channels=8*64,out_channels=64*4, kernel_size=2, stride=2, padding=0)
    self.conv31 = nn.Conv2d(in_channels=8*64, out_channels=64*4, kernel_size=4, stride = 1,padding=2)
    self.conv32 = nn.Conv2d(in_channels=4*64, out_channels=64*4, kernel_size=4, stride = 1,padding=1)
    self.convTran3 = nn.ConvTranspose2d(in_channels=4*64,out_channels=64*2, kernel_size=2, stride=2, padding=0)
    self.conv41 = nn.Conv2d(in_channels=4*64, out_channels=64*2, kernel_size=4, stride = 1,padding=2)
    self.conv42 = nn.Conv2d(in_channels=2*64, out_channels=64*2, kernel_size=4, stride = 1,padding=1)
    self.convTran4 = nn.ConvTranspose2d(in_channels=2*64,out_channels=64, kernel_size=2, stride=2, padding=0)
    self.conv51 = nn.Conv2d(in_channels=2*64, out_channels=64, kernel_size=4, stride = 1,padding=1)
    self.conv52 = nn.Conv2d(in_channels=64, out_channels=3, kernel_size=4, stride = 1,padding=2)
    
   
  def forward(self, latent, capas):
    #out = self.fc(latent)
    #out = out.view(out.size(0), 64*2, 64, 64)
    #out_encoder = capas[3]
    out = latent
    #out = torch.cat((latent,out_encoder), dim=1)
    #out = torch.cat((out,out_encoder), dim=1)
    #print( out_encoder.size(), out.size())
    #out = out.view(out.size(0), 128*2, 64, 64)
    out = F.relu(self.conv11(latent))
    out = self.convTran1(out)
    #print("Convtrav 1",out.size())
    out = torch.cat((out,capas[0]), dim=1)
    out = F.relu(self.conv21(out))
    out = F.relu(self.conv22(out))
    out = F.relu(self.convTran2(out))
    #print("Convtrav 2",out.size())
    out = torch.cat((out,capas[1]), dim=1)
    out = F.relu(self.conv31(out))
    out = F.relu(self.conv32(out))
    out = F.relu(self.convTran3(out))
    #print("Convtrav 3",out.size())
    out = torch.cat((out,capas[2]), dim=1)
    out = F.relu(self.conv41(out))
    out = F.relu(self.conv42(out))
    out = F.relu(self.convTran4(out))
    #print("Convtrav 4",out.size())
    out = torch.cat((out,capas[3]), dim=1)
    out = F.relu(self.conv51(out))
    out = F.relu(self.conv52(out))
    return out
```

```python
class Autoencoder(nn.Module):
   def __init__(self):
        super(Autoencoder, self).__init__()
        self.encoder = Encoder()
        self.decoder = Decoder()

   def forward(self, x):
        latent, capas = self.encoder(x)
        x_recon = self.decoder(latent, capas)
        return  x_recon
```

```python
def validation(model, valid_loader_lowres, valid_loader_highres, loss_fn):
    model.eval()
    loss_total = 0

    # Test validation data
    with torch.no_grad():
        for item1, item2 in zip(valid_loader_lowres, cycle(valid_loader_highres)):
            image_batch, _ = item1
            image_compare, _ = item2
            image_batch = image_batch.to(device)
            image_compare = image_compare.to(device)
            image_batch_recon = model(image_batch)
            loss = loss_fn(image_batch_recon, image_compare)
            loss_total += loss.item()

    return loss_total / len(valid_loader_highres)
```

## Aprendizaje de modelo

```python
from itertools import cycle

def train(model, Epochs, loss_fn, train_loader_lowres, train_loader_highres,valid_loader_lowres, valid_loader_highres):
    train_loss_avg = []
    valid_loss_avg = []
    for epoch in range(Epochs):
      train_loss_avg.append(0)
      num_batches = 0
    
      for item1, item2 in zip(train_loader_lowres, cycle(train_loader_highres)):
          image_batch, _ = item1
          image_compare, _ = item2

          image_batch = image_batch.to(device)
          image_compare = image_compare.to(device)
          
          image_batch_recon = model(image_batch)
          loss = loss_fn(image_batch_recon, image_compare)
          
          optimizer.zero_grad()
          loss.backward()
          optimizer.step()
          
          train_loss_avg[-1] += loss.item()
          num_batches += 1
          
      train_loss_avg[-1] /= num_batches
      print('Epoch [%d / %d] average reconstruction error: %f' % (epoch+1, Epochs, train_loss_avg[-1]))
      current_loss = validation(model, valid_loader_lowres, valid_loader_highres, loss_fn)
      valid_loss_avg.append(current_loss)
    return train_loss_avg, valid_loss_avg
```
